```python
import pandas as pd
import numpy as np
import seaborn as sns
```

Приводим таблицы в адекватный вид, с нужной информацией

```python
#Загрузка данных
train_path = 'train.csv'
valid_path = 'valid.csv'

train_data = pd.read_csv(train_path)
valid_data = pd.read_csv(valid_path)

# Очистка данных
# Удаление дублирующихся строк
train_cleaned = train_data.drop_duplicates()
valid_cleaned = valid_data.drop_duplicates()

# Заполнение пропущенных значений
# Для числовых столбцов заполняем -1
numeric_cols_train = train_cleaned.select_dtypes(include=[np.number]).columns
numeric_cols_valid = valid_cleaned.select_dtypes(include=[np.number]).columns

train_cleaned[numeric_cols_train] = train_cleaned[numeric_cols_train].apply(pd.to_numeric, errors='coerce')
valid_cleaned[numeric_cols_valid] = valid_cleaned[numeric_cols_valid].apply(pd.to_numeric, errors='coerce')

train_cleaned[numeric_cols_train] = train_cleaned[numeric_cols_train].fillna(-1)
valid_cleaned[numeric_cols_valid] = valid_cleaned[numeric_cols_valid].fillna(-1)

# Для категориальных столбцов заполняем модой
categorical_cols_train = train_cleaned.select_dtypes(include=['object']).columns
categorical_cols_valid = valid_cleaned.select_dtypes(include=['object']).columns

train_cleaned[categorical_cols_train] = train_cleaned[categorical_cols_train].fillna(train_cleaned[categorical_cols_train].mode().iloc[0])
valid_cleaned[categorical_cols_valid] = valid_cleaned[categorical_cols_valid].fillna(valid_cleaned[categorical_cols_valid].mode().iloc[0])

# Замена повторяющихся значений в строках на -1 (кроме первого вхождения)
def replace_duplicates_with_minus_one(row):
    seen = set()
    return row.map(lambda x: x if x not in seen and not seen.add(x) else -1)

train_cleaned = train_cleaned.apply(replace_duplicates_with_minus_one, axis=1)
valid_cleaned = valid_cleaned.apply(replace_duplicates_with_minus_one, axis=1)

# Удаление колонок, где более 90% значений равны -1
threshold_train = len(train_cleaned) * 0.9
threshold_valid = len(valid_cleaned) * 0.9

cols_to_remove_train = [col for col in train_cleaned.columns if (train_cleaned[col] == -1).sum() > threshold_train]
cols_to_remove_valid = [col for col in valid_cleaned.columns if (valid_cleaned[col] == -1).sum() > threshold_valid]

train_cleaned = train_cleaned.drop(columns=cols_to_remove_train)
valid_cleaned = valid_cleaned.drop(columns=cols_to_remove_valid)

# Удаление дублирующихся строк
train_cleaned = train_cleaned.drop_duplicates()
valid_cleaned = valid_cleaned.drop_duplicates()

# Вывод количества и названий оставшихся столбцов
print(f"Количество оставшихся столбцов в обучающих данных: {train_cleaned.shape[1]}")
print(f"Столбцы в обучающих данных: {list(train_cleaned.columns)}")
print(f"Количество оставшихся столбцов в валидационных данных: {valid_cleaned.shape[1]}")
print(f"Столбцы в валидационных данных: {list(valid_cleaned.columns)}")

# Сохранение очищенных данных
train_cleaned.to_csv('cleaned_train.csv', index=False)
valid_cleaned.to_csv('cleaned_valid.csv', index=False)
```

Импорт для моделей

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score
```

Для train

```python
# Загрузка тренировочных данных
train_data = pd.read_csv('cleaned_train.csv')

# Разделение признаков и целевой переменной
X_train = train_data.drop(columns=['client_id', 'target'])
y_train = train_data['target']

# Преобразование категориальных признаков
X_train = pd.get_dummies(X_train, drop_first=True)

# Обучение модели Logistic Regression
model = LogisticRegression(random_state=42, class_weight='balanced', max_iter=500)
model.fit(X_train, y_train)

# Оценка качества модели на тренировочной выборке
y_pred_train = model.predict(X_train)
print("Отчет на обучающей выборке:")
print(classification_report(y_train, y_pred_train))

roc_auc_train = roc_auc_score(y_train, model.predict_proba(X_train)[:, 1])
print(f"ROC-AUC на обучающей выборке: {roc_auc_train:.3f}")

# Предсказание вероятностей
y_proba_train = model.predict_proba(X_train)[:, 1]

# Фильтрация потенциальных покупателей (вероятность >= 0.6)
threshold = 0.6
potential_buyers_train = train_data[y_proba_train >= threshold]

# Сохранение потенциальных покупателей в файл
potential_buyers_train.to_csv('potential_buyers_train.csv', index=False)
print(f"Сохранено {len(potential_buyers_train)} потенциальных покупателей в файл 'potential_buyers_train.csv'.")
```

Для valid

```python
# Загрузка тренировочных данных (для обучения модели) и валидационной выборки
train_data = pd.read_csv('cleaned_train.csv')
valid_data = pd.read_csv('cleaned_valid.csv')

# Разделение признаков и целевой переменной для тренировки
X_train = train_data.drop(columns=['client_id', 'target'])
y_train = train_data['target']

# Разделение признаков и целевой переменной для валидации
X_valid = valid_data.drop(columns=['client_id', 'target'])
y_valid = valid_data['target']

# Преобразование категориальных признаков
X_train = pd.get_dummies(X_train, drop_first=True)
X_valid = pd.get_dummies(X_valid, drop_first=True)

# Выравнивание столбцов валидационных данных с тренировочными
X_valid = X_valid.reindex(columns=X_train.columns, fill_value=0)

# Обучение модели Logistic Regression
model = LogisticRegression(random_state=42, class_weight='balanced', max_iter=500)
model.fit(X_train, y_train)

# Оценка качества модели на валидационной выборке
y_pred_valid = model.predict(X_valid)
print("Отчет на валидационной выборке:")
print(classification_report(y_valid, y_pred_valid))

roc_auc_valid = roc_auc_score(y_valid, model.predict_proba(X_valid)[:, 1])
print(f"ROC-AUC на валидационной выборке: {roc_auc_valid:.3f}")

# Предсказание вероятностей
y_proba_valid = model.predict_proba(X_valid)[:, 1]

# Фильтрация потенциальных покупателей (вероятность >= 0.6)
threshold = 0.6
potential_buyers_valid = valid_data[y_proba_valid >= threshold]

# Сохранение потенциальных покупателей в файл
if not potential_buyers_valid.empty:
    potential_buyers_valid.to_csv('potential_buyers_valid.csv', index=False)
    print(f"Сохранено {len(potential_buyers_valid)} потенциальных покупателей в файл 'potential_buyers_valid.csv'.")
else:
    print(f"Нет покупателей с вероятностью >= {threshold}.")
```

Матрицы до очистки/после очистки

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Загрузка данных
train_data = pd.read_csv('train.csv')  # Исходные данные до очистки
cleaned_train_data = pd.read_csv('cleaned_train.csv')  # Очищенные данные

valid_data = pd.read_csv('valid.csv')  # Исходные данные до очистки
cleaned_valid_data = pd.read_csv('cleaned_valid.csv')  # Очищенные данные

# Функция для визуализации пропусков
def plot_missing_matrix(data, title):
    plt.figure(figsize=(10, 6))
    sns.heatmap(data.isnull(), cbar=False, cmap='viridis', yticklabels=False)
    plt.title(title)
    plt.xlabel("Признаки")
    plt.ylabel("Записи")
    plt.show()

# Тренировочная выборка: до и после очистки
plot_missing_matrix(train_data, "Матрица пропусков (Тренировочная выборка, до очистки)")
plot_missing_matrix(cleaned_train_data, "Матрица пропусков (Тренировочная выборка, после очистки)")

# Валидационная выборка: до и после очистки
plot_missing_matrix(valid_data, "Матрица пропусков (Валидационная выборка, до очистки)")
plot_missing_matrix(cleaned_valid_data, "Матрица пропусков (Валидационная выборка, после очистки)")
```

ROC-AUC графики

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Загрузка очищенных данных
cleaned_train_data = pd.read_csv('cleaned_train.csv')
cleaned_valid_data = pd.read_csv('cleaned_valid.csv')

# Разделение признаков и целевой переменной
X_train = cleaned_train_data.drop(columns=['client_id', 'target'])
y_train = cleaned_train_data['target']
X_train = pd.get_dummies(X_train, drop_first=True)

X_valid = cleaned_valid_data.drop(columns=['client_id', 'target'])
y_valid = cleaned_valid_data['target']
X_valid = pd.get_dummies(X_valid, drop_first=True)
X_valid = X_valid.reindex(columns=X_train.columns, fill_value=0)

# Предсказание вероятностей для ROC-AUC
y_proba_train = model.predict_proba(X_train)[:, 1]
y_proba_valid = model.predict_proba(X_valid)[:, 1]

# Функция для построения ROC-кривой
def plot_roc_curve(y_true, y_proba, title):
    fpr, tpr, _ = roc_curve(y_true, y_proba)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC-AUC = {roc_auc:.3f}')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(title)
    plt.legend(loc="lower right")
    plt.grid()
    plt.show()

# Построение ROC-кривой для тренировочной выборки
plot_roc_curve(y_train, y_proba_train, "ROC-AUC - Тренировочная выборка")

# Построение ROC-кривой для валидационной выборки
plot_roc_curve(y_valid, y_proba_valid, "ROC-AUC - Валидационная выборка")
```
